```python
import numpy as np
from matplotlib import pyplot as plt
import ase.units as au
```

```python
def get_minimum(x,y):
    par = np.polyfit(x, y, deg = 2)
    x_min = -par[1]/(2*par[0])
    return(x_min)
```
