Expanding the first derivative we get $f'(x_0+dx)=f'(x_0)+f''(x_0)dX+\frac{1}{2}f'''(x_0)dx^2$
In order to get a approximate value for the minimum assuming the derivative is zero we have to solve for dX:
$\frac{1}{2}f'''(x_0)dx^2+f''(x_0)dx+f'(x_0)=0$  
$ dX=\frac{-f'' \pm \sqrt{f''^2 -2f'f'''}}{f'''} $
thus:

```python
dX=(-fd2+(fd2**2-2.*fd1*fd3)**0.5)/fd3
X_pr=2.025+dX
print dX ,X_pr
dX2=-fd1/fd2
print dX2,2.025+dX2
```

```python
def FourPointPrediction(x,h):
   # '''a prediction of the minimum of a function based on a 4 point approx'''
    points =[]
    for i in range (4):
        points.append(uhf(x+h*(i-1.5),[6.,8.])[0])
    print points
    fd3=(points[3]-3*points[2]+3*points[1]-points[0])/(h**3)
    fd1=(points[2]-points[1]-h**3/24*fd3)/h
    fd2=(points[3]-points[2]-points[1]+points[0])/(2*h**2)
    
    print 'the derivatives are:', fd1,fd2,fd3,'to cm:' ,to_cm(fd2,Mu_CO)
    #predictions
    delta=fd2**2-2.*fd1*fd3
    print 'delta =', delta
    result=[x]
    dX2=-fd1/fd2
    X_pr2=x+dX2
    d_pr2=fd2+fd3*dX2
    print X_pr2, to_cm(d_pr2,Mu_CO),'first order approx X_pr / wn_pr'
    result.append([X_pr2, to_cm(d_pr2,Mu_CO)])
    
    if delta>0:
        dX=(-fd2+(delta)**0.5)/fd3
        X_pr=x+dX
        d_pr=fd2+fd3*dX
        print X_pr, to_cm(d_pr,Mu_CO),'second order approx X_pr / wn_pr'
        result.append([X_pr, to_cm(d_pr,Mu_CO)])
    return result
```

```python
results=[]
for x in np.linspace(1.5,2.5,11):
    results.append(FourPointPrediction(x,.001))
```

```python
plt.plot([x[0] for x in results],[x[1][1] for x in results])
for x in results: print x[0],x[1][1]
plt.axhline(2424.405, color="red")
plt.show()
plt.plot([x[0] for x in results],[x[1][0] for x in results])
for x in results: print x[0],x[1][0]
plt.axhline(2.08711, color="red")
plt.show()
```
