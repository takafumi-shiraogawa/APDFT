```python
from horton import *
import numpy as np
import matplotlib.pyplot as plt 
from optimiz import optimiz
from NNfromCO import toNN
from co_scf_ToNN import uhf,b3lyp
def to_cm(k,Mu):
    return (k/Mu)**0.5*centimeter/planck/lightspeed
Mu_CO=periodic['c'].mass*periodic['o'].mass/(periodic['c'].mass+periodic['o'].mass)
Mu_NN=periodic['n'].mass/2.
```

```python
h=0.05
points =[]
for i in range (4):
    points.append(toNN(2.127+h*(i-1.5),scf=b3lyp))

fd3=(points[3]-3*points[2]+3*points[1]-points[0])/(h**3)
fd1=(points[2]-points[1]-h**3/24*fd3)/h
fd2=(points[3]-points[2]-points[1]+points[0])/(2*h**2)

print 'the derivatives are:', fd1,fd2,fd3,'to cm:' ,to_cm(fd2,Mu_NN)
#predictions

result=[]
dX2=-fd1/fd2
X_pr2=x+dX2
d_pr2=fd2+fd3*dX2
print X_pr2, to_cm(d_pr2,Mu_NN),'first order approx X_pr / wn_pr'
result.append([X_pr2, to_cm(d_pr2,Mu_NN)])

delta=fd2**2-2.*fd1*fd3
print 'delta =', delta
if delta>0:
    dX=(-fd2+(delta)**0.5)/fd3
    X_pr=x+dX
    d_pr=fd2+fd3*dX
    print X_pr, to_cm(d_pr,Mu_NN),'second order approx X_pr / wn_pr'
    result.append([X_pr, to_cm(d_pr,Mu_NN)])
```
