```python
from matplotlib import pyplot as plt
plt.rcParams['font.size'] = 20
from matplotlib import cm as cmx
import matplotlib.colors as colors
import numpy as np
```

```python
atomic_all = np.loadtxt('/home/misa/APDFT/prototyping/atomic_energies/results/analyse_learning/lcurves_atomic/cm/best_all_atomic.txt')
all_alchpots = np.loadtxt('/home/misa/APDFT/prototyping/atomic_energies/results/analyse_learning/lcurves_alch_pot/best_all_alchpots.txt')
```

```python
datasets = [all_alchpots, atomic_all]
labels = [r'$\mu$', r'$E_I$']
```

```python
fig, ax = plt.subplots(1,1)
plt.rcParams.update({'font.size': 18})

for d,l in zip(datasets, labels):
    ax.errorbar(d[:,0], d[:,1], d[:,2], label=l)

ax.set_xlabel('# Training points')
ax.set_ylabel('MAE (Ha)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()
# ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
directory = '/home/misa/projects/Atomic-Energies/figures/machine_learning/learning_curves/'
plt.savefig(directory+'all_alchpot_atomic.png', bbox_inches='tight')
```
