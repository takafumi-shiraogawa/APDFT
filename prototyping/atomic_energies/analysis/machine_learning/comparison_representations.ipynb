### learning all elements together

```python
datasets = [all_alchpots_cm, all_alchpots_aslatm, all_alchpots_fchl]
labels = [r'all $\mu$ CM', r'all $\mu$ aSLATM', r'all $\mu$ FCHL']

plt.rcParams.update({'font.size': 18})
fig, ax = plt.subplots(1,1)
# plt.rcParams.update({'errorbar.capsize': 0})

for d,l in zip(datasets, labels):
    ax.errorbar(d[:,0], d[:,1], d[:,2], label=l)

ax.set_xlabel('$N$')
ax.set_ylabel('MAE (Ha)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
directory = '/home/misa/projects/Atomic-Energies/figures/machine_learning/learning_curves/'
plt.savefig(directory+'lcurve_all_mu_cm_aslatm_fchl.png', bbox_inches = 'tight')
```
