```python
import matplotlib
matplotlib.use('Qt5Agg')
import numpy as np
from matplotlib import pyplot as plt
```

```python
ve = 38
ve_new = np.linspace(0, 38, 39)
lamb = ve_new/ve

plt.rcParams.update({'font.size': 18})
plt.plot(ve_new, lamb, 'o')
plt.xlabel('# valence electrons')
plt.ylabel(r'$\lambda$-value')
```
