```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

plt.style.use("/mnt/c/Users/guido/workcopies/dotfiles/plotstyle/col1.mplstyle")
```

```python
def plot_image_to_figure(figure, filename):
    im = Image.open(filename)
    height = im.size[1]
    im = np.array(im).astype(np.float) / 255
    figure.figimage(im, 60, 120, zorder=10)
    
def do_schematic():
    f = plt.figure(figsize=(5,5))
    plt.plot((5,9), (5,9), color='grey')
    #plt.plot((5, 8), (6,9), color='lightgrey')
    plt.xlabel('$Z_I$')
    plt.ylabel('$Z_J$')
    
    #N2, CO
    plt.scatter((6,8), (8, 6), color='C0', s=40)
    plt.scatter((7,), (7,), color='white', facecolor='white', s=120, zorder=19)
    plt.scatter((7,), (7,), color='C0', facecolor='none', zorder=20, s=40)
    #plt.annotate(s='N$_2$', xy=(7,7), xytext=(7, 6.8),ha='center', va='top')
    #plt.annotate(s='CO', xy=(6,8), xytext=(6, 7.8),ha='center', va='top')
    #plt.annotate(s='OC', xy=(8,6), xytext=(8, 5.8),ha='center', va='top')
    plt.annotate("", xy=(6, 8), xytext=(8, 6), arrowprops=dict(arrowstyle="<|-|>", facecolor='grey', edgecolor='grey',shrinkA=5, shrinkB=5), zorder=-100)
    
    
    # BF
    plt.scatter((5,9), (9, 5), color='C0', s=40)
    plt.annotate("", xy=(8, 6), xytext=(9, 5), arrowprops=dict(arrowstyle="<|-", facecolor='grey',edgecolor='grey', shrinkA=5, shrinkB=8))
    plt.annotate("", xy=(6, 8), xytext=(5, 9), arrowprops=dict(arrowstyle="<|-", facecolor='grey',edgecolor='grey', shrinkA=5, shrinkB=8))
    
    # CN-
    plt.scatter((6,7), (7, 6), color='C1', s=40, marker='^')
    plt.scatter((6.5,), (6.5,), color='white', facecolor='white', s=120, zorder=19, marker='^')
    plt.scatter((6.5,), (6.5,), color='C1', facecolor='none', zorder=20, s=40, marker='^')
    plt.annotate("", xy=(6, 7), xytext=(7, 6), arrowprops=dict(arrowstyle="<|-|>", facecolor='grey',edgecolor='grey', shrinkA=5, shrinkB=10))
    plt.annotate("virtual midpoint", xy=(6.5, 6.5), xytext=(5.8, 5.5), arrowprops=dict(arrowstyle="-|>", facecolor='black', shrinkB=10))

    # invalid
    plt.scatter((7,9), (8, 8), color='C2', s=40, marker='v')
    plt.annotate("", xy=(7, 8), xytext=(9, 8), arrowprops=dict(arrowstyle="<|-|>", facecolor='grey', edgecolor='grey',shrinkA=5, shrinkB=10, linewidth=0))
    plt.plot((7.2,8.9), (8,8), linestyle='--', dashes=(5,3), color='grey', zorder=-100)
    plt.annotate("no chirality", xy=(8, 8), xytext=(6.8, 8.5), arrowprops=dict(arrowstyle="-|>", facecolor='black', shrinkB=10))
    
    # derivatives
    #plt.annotate("", xy=(7, 7), xytext=(7, 7.5), arrowprops=dict(arrowstyle="<|-", facecolor='black', shrinkB=10))
    #plt.annotate(r"$\int_\Omega \Delta V \frac{1}{n!}\partial^{n-1}_{Z_I}\rho$", xy=(7, 7), xytext=(7.5, 7), arrowprops=dict(arrowstyle="<|-", facecolor='black', shrinkB=10))
    
    plt.annotate(r"$\tilde\sigma_a$", xy=(8.5, 9))

    plot_image_to_figure(f, 'py2.png')
    
    plt.savefig('schematic.pdf', bbox_inches='tight')
do_schematic()
```
